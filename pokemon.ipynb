# Regex filtering

```python
import re
pokemon.loc[pokemon['Type 1'].str.contains('fire|grass, regex=True')] #the code doesnt give any data due to error in capitalization
```

```python
pokemon.loc[pokemon['Type 1'].str.contains('fire|grass', regex=True, flags=re.I)] #Here flags=re.I helps to remove ignore the capitalization
```

```python
pokemon.loc[pokemon['Type 1'].str.contains('fire|grass', regex=True, flags=re.I)]
```

```python
pokemon.loc[pokemon['Type 1'].str.contains('poision|water', regex=True, flags=re.I)]
```

To get dataset thats contains cetain words then we can use re

```python
import re
#pokemon.loc[pokemon['Name'].str.contains('pi'[a-z])] #It gives all the name that has pi, if we use "Pi" then we will get Pi
pokemon.loc[pokemon['Name'].str.contains('^Pi[a-z]*', regex=True, flags=re.I)] 

import re
pokemon.loc[pokemon['Name'].str.contains('mega[a-z]*', regex=True, flags=re.I)]

pokemon.loc[pokemon['Type 1'].str.contains('Gra[a-z]*', regex=True, flags=re.I)]
```
