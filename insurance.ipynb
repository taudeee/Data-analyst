#count of claims per agent and vendor

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual name of your DataFrame

# Count of claims per agent and vendor
plt.figure(figsize=(14, 8))
sns.countplot(x='AGENT_ID', hue='CLAIM_STATUS', data=di, palette='viridis')
plt.title('Claims Handling by Agents')
plt.xlabel('Agent ID')
plt.ylabel('Count')
plt.legend(title='Claim Status', loc='upper right')
plt.show()

plt.figure(figsize=(14, 8))
sns.countplot(x='VENDOR_ID', hue='CLAIM_STATUS', data=di, palette='muted')
plt.title('Claims Handling by Vendors')
plt.xlabel('Vendor ID')
plt.ylabel('Count')
plt.legend(title='Claim Status', loc='upper right')
plt.show()
```

#incident severity and reporting

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Replace 'df' with the actual name of your DataFrame

# Count of incidents by severity
plt.figure(figsize=(10, 6))
sns.countplot(x='INCIDENT_SEVERITY', data=di, palette='rocket')
plt.title('Distribution of Incident Severities')
plt.xlabel('Incident Severity')
plt.ylabel('Count')
plt.show()
```

```python
# Grouped bar plot for patterns
plt.figure(figsize=(14, 8))

# Authority Contacted
plt.subplot(1, 3, 1)
sns.countplot(x='INCIDENT_SEVERITY', hue='AUTHORITY_CONTACTED', data=di, palette='viridis')
plt.title('Incident Severity vs Authority Contacted')
plt.xlabel('Incident Severity')
plt.ylabel('Count')
plt.legend(title='Authority Contacted', loc='upper right')

# Any Injury
plt.subplot(1, 3, 2)
sns.countplot(x='INCIDENT_SEVERITY', hue='ANY_INJURY', data=di, palette='muted')
plt.title('Incident Severity vs Any Injury')
plt.xlabel('Incident Severity')
plt.ylabel('Count')
plt.legend(title='Any Injury', loc='upper right')

# Police Report Available
plt.subplot(1, 3, 3)
sns.countplot(x='INCIDENT_SEVERITY', hue='POLICE_REPORT_AVAILABLE', data=di, palette='pastel')
plt.title('Incident Severity vs Police Report Available')
plt.xlabel('Incident Severity')
plt.ylabel('Count')
plt.legend(title='Police Report Available', loc='upper right')

plt.tight_layout()
plt.show()
```
