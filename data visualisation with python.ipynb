```python
countries = pd.read_csv("countries.csv")
```

```python
countries
```

```python
countries.country
```

```python
#compare the population in china and us
```

```python
countries.country == "United States" #this shows the population of us in different column
```

```python
countries[countries.country == "United States"] #this show the population of us 
```

```python
us = countries[countries.country == "United States"]
```

```python
china = countries[countries.country == "China"]
```

```python
us
```

```python
china
```

```python
plt.plot(us.year, us.population / 10 ** 6)
plt.plot(china.year, china.population / 10 ** 6)
plt.title("comparision between U.S and China in population")
plt.xlabel("year in A.D")
plt.ylabel("population in millions")
plt.legend(["U.S", "china"])
plt.show()
```

```python
#percentage growth of U.S and China
```

```python
us.population
```

```python
us.population.iloc[0]
```

```python
us.population / us.population.iloc[0] * 100 
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.title("comparision between U.S and China in population")
plt.xlabel("year in A.D")
plt.ylabel("population in millions")
plt.legend(["U.S", "china"])
plt.show()
```
